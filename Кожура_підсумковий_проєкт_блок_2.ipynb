**Завдання 2. Аналіз повноти відповідей респондентів**

```python
# Отримуємо список питань із schema
# колонка qname містить назви всіх питань. questions - створюємо множину назв питань, які присутні одночасно в qname та серед стовпців df.
# Використання set() та qname.unique() створює множину унікальних qname
questions = set(df1['qname'].unique())
# len(questions)
print(f"Загальна кількість унікальних питань: {len(questions)}")
```

```python
# Беремо лише ті питання, які реально є в основному датасеті:
# ми беремо перетин із df.columns, щоб отримати лише ті питання, які реально існують у датафреймі
questions_in_df = list(questions.intersection(set(df.columns)))
# len (questions_in_df)
print(f"Кількість питань, які реально є в основному датасеті: {len (questions_in_df)}")
```

```python
# Видаляємо респондентів, у яких є пропуски хоча б в одному з цих питань
# dropna(): видаляє рядки, де є пропуски (NaN) у зазначених стовпцях.
answered_all = df[questions_in_df].dropna()
```

```python
# Рахуємо кількість таких респондентів
# shape[0] повертає кількість рядків-тобто кількість респондентів, які відповіли на всі ці питання.
total_full_answers = answered_all.shape[0]

print(f"Кількість респондентів, які відповіли на всі ці питання: {total_full_answers}")
```

**Завдання 5. Визначення популярності Python**

```python
#Знаходимо правильну колонку
# У stack-overflow-developer-survey-2025 мови програмування повинні бути у колонці:
# LanguageHaveWorkedWith
# Дивимося, як виглядають значення:
df['LanguageHaveWorkedWith'].head (10)
```

```python
# Створюємо новий стовпець worked_with_python, який містить True/False, 
# де True означає, що респондент згадував Python у LanguageHaveWorkedWith.
# str.contains('...') - шукаємо слово 'python' 
# без урахування регістру та замінюємо пропуски (NaN) на False.
df['worked_with_python'] = ( 
    df['LanguageHaveWorkedWith']
    .str.contains('python', case=False, na=False)
)
df['worked_with_python'].head (10)
```

```python
# sum() - рахуємо загальну кількість респондентів, які працювали з Python.
python_devs = df['worked_with_python'].sum()
# python_devs
# виводимо результат:
print(f"Загальна кількість Python-розробників: {python_devs}")
```

```python
# - обчислюємо частку таких респондентів у відсотках.
total_respondents = df.shape[0]
python_percent = (python_devs / total_respondents) * 100
```

```python
# виводимо результат:
print(f"Відсоток респондентів, які програмують на Python: {python_percent:.1f}%")
```

**Завдання 7. Географічний аналіз компенсації Python-розробників**

```python
# df.worked_with_python - відбираємо лише тих, хто працював з Python.
python_df = df[df['worked_with_python']].copy()
```

```python
# Переконуємось, що зарплата числова
python_df['ConvertedCompYearly'] = pd.to_numeric(
    python_df['ConvertedCompYearly'], errors='coerce'
)
python_df['ConvertedCompYearly']
```

```python
# Видаляємо рядки, де відсутнє значення зарплати (ConvertedCompYearly).
python_df = python_df.dropna(subset=['ConvertedCompYearly','Country'])
```

```python
# .groupby() - групуємо дані за країною (Country)
# і обчислюємо середню (mean) та медіанну (median) зарплату для кожної країни.
# py_progers_by_country - таблиця з резульататами.
py_progers_by_country = (
    python_df.groupby('Country')['ConvertedCompYearly'] 
    .agg(['mean', 'median']).round (2) 
    .reset_index()
)
# Зробили назви колонок зрозумілими
py_progers_by_country.columns = [
    'Country',
    'Mean_Compensation_Yearly',
    'Median_Compensation_Yearly'
]
py_progers_by_country.head()
```

**Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів**

```python
# Вибираємо три стовпці: ResponseId, EdLevel, ConvertedCompYearly.
top5_education = (
    df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]  
      .copy()
)
# Переконуємося, що зарплата - число.
top5_education['ConvertedCompYearly'] = pd.to_numeric(    
    top5_education['ConvertedCompYearly'], errors='coerce'
)
# Прибираємо пропуски зарплати, сортуємо у порядку спадання, оновлюємо індекс після сотрування.
# Відбираємо перші 5 рядків (топ-5 респондентів з найвищою зарплатою).
top5_education = (
    top5_education
      .dropna(subset=['ConvertedCompYearly'])              
      .sort_values(by='ConvertedCompYearly', ascending=False)  
      .reset_index(drop=True)                               
      .head(5)                                              
)
# Виводимо таблицю з 5 респондентами з найбільшою зарплатою і їх:
# ResponseId, EdLevel — рівень освіти, ConvertedCompYearly — річна компенсація
top5_education[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]
```

**Завдання 9. Аналіз популярності Python по віковим категоріям**

```python
# Беремо колонку з віком: Age
# Дивимося унікальні значення:
df['Age'].unique()
```

```python
# Створюємо колонку "чи використовує Python"
df['UsesPython'] = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)
```

```python
python_by_age = (
    df.groupby('Age')['UsesPython'] # Групування по віку і розрахунок %
    .mean()
    .reset_index()
    .assign(PythonPercent = lambda x: (x['UsesPython'] * 100).round(1)) # Перевели у відсотки
    [['Age', 'PythonPercent']] # Залишаємо потрібні колонки
    .sort_values('PythonPercent', ascending=False) # Сортуємо та дивимось результат
)

python_by_age.head()
```

**Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників**

```python
# Прибраємо пропуски в зарплаті
# Перцентиль рахується тільки по реальних значеннях.
comp_df = df.dropna(subset=['ConvertedCompYearly'])
```

```python
# Обчислюємо 75-й перцентиль
# Це поріг: вище нього = високооплачувані
p75 = comp_df['ConvertedCompYearly'].quantile(0.75)
print(f"75-й перцентиль дорівнює:  {p75}")
```

```python
# Відфільтрувуємо високооплачуваних
high_paid_df = comp_df[comp_df['ConvertedCompYearly'] > p75]
```

```python
# Дивимося колонку RemoteWork
# Перевіряємо значення:
df['RemoteWork'].unique()
```

```python
# Віддалена робота містить текст: "Remote".
# Відфільтрувуємо тих, хто працює віддалено:
remote_high_paid_df = high_paid_df[
    high_paid_df['RemoteWork'].str.contains('Remote', na=False)
]
remote_high_paid_df.head()
```

```python
# Дивимось колонку з індустріями: Industry
# Перевіряємо:
# df[['Industry']].unique()
df['Industry'].value_counts(dropna=False)
```

```python
# Знайдемо найпопулярніші індустрії
top_industries = (
    remote_high_paid_df['Industry']
    .dropna()
    .value_counts()
    .reset_index()
)
top_industries.columns = ['Industry', 'Count']
top_industries
```
